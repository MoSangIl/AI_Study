# H(x) = Wx + b

```python
import tensorflow.compat.v1 as tf

# tf.disable_v2_behavior()
```

```python
# X and Y data
# x_train = [1,2,3]
# y_train = [1,2,3]

## Using Placeholder
tf.disable_eager_execution()
X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])

# tensorflow 에 학습시킬 변수를 설정 하는 것 trainable value
## initial value 는 랜덤 값으로, [] -> shape
W = tf.Variable(tf.random.normal([1]), name='weight')
b = tf.Variable(tf.random.normal([1]), name='bias')

# Hypothesis
hypothesis = W * X + b
```

```python
# cost / Loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
# reduce mean -> 주어진 tensor  값들의 평균값을 내준다.
print(cost)
```

```python
# Minimize !! -> GradientDescent
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```python
# Launch the graph ion a session
sess = tf.Session()
# Initializes global variables (W, b) in the graph 
sess.run(tf.global_variables_initializer())

# Fit the line
for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict={X: [1, 2, 3, 4, 5], Y:[1, 2, 3, 4, 5]})
    ## 20번에 한번씩 값을 출력해보자.
    if step % 20 == 0:
#         print(step, sess.run(cost), sess.run(W), sess.run(b))
# placeholder 사용시에는, X, Y 값이 없기 때문에 바로 값을 출력 할 수 없음
        print(step, cost_val, W_val, b_val)
```

```python
# Testing our model
print(sess.run(hypothesis, feed_dict={X: [5]}))
print(sess.run(hypothesis, feed_dict={X: [2.5]}))
print(sess.run(hypothesis, feed_dict={X: [1.5, 3.5]}))
```
